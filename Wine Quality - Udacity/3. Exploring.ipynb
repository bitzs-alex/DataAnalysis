```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
wine = pd.read_csv('./Datasets/combined_qualities.csv')
```

```python
wine.shape
```

## Exploring using Diagrams

```python
fig, (ax_1, ax_2) = plt.subplots(nrows=2, ncols=2, figsize=(14, 14))
ax_1[0].hist(wine.fixed_acidity)
ax_1[0].set_xlabel('Fixed Acidity')

ax_1[1].hist(wine.total_sulfur_dioxide)
ax_1[1].set_xlabel('Total Sulfur Dioxide')

ax_2[0].hist(wine.ph)
ax_2[0].set_xlabel('PH')

ax_2[1].hist(wine.alcohol)
ax_2[1].set_xlabel('Alcohol Level');
```

```python
fig, (ax_1, ax_2) = plt.subplots(nrows=2, ncols=2, figsize=(14, 14))
ax_1[0].scatter(wine.quality, wine.volatile_acidity)
ax_1[0].set_xlabel('Quality vs Volatile Acidity')

ax_1[1].scatter(wine.quality, wine.residual_sugar)
ax_1[1].set_xlabel('Quality vs Residual Sugar')

ax_2[0].scatter(wine.quality, wine.ph)
ax_2[0].set_xlabel('Quality vs PH')

ax_2[1].scatter(wine.quality, wine.alcohol)
ax_2[1].set_xlabel('Quality vs Alcohol Level');
```
